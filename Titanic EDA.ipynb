```python
import pandas as pd
import seaborn as sns

#importing training dataset
df = pd.read_csv('train.csv')

#counting values of  survived people
df["Survived"].value_counts()

#No of null value in df dataset
df.isnull().value_counts()

#Dropping Passenger name and ID as they are unrelated to survival
df.drop(['PassengerId','Name',], axis=1, inplace=True)

#Filling out missing values of Age column
df['Age'].fillna(df['Age'].median(), inplace=True)

#
df.loc[df['Age'] < 16, 'Sex'] = 'Child'

#Changing data type from object to category and float
df['Embarked'] = df['Embarked'].astype('category')
df['Sex'] = df['Sex'].astype('category')
df['Fare'] = df['Fare'].astype(float)

#Data Visualization

df.Survived.hist()
sns.catplot(x='Sex', data=df, hue='Survived',kind='count')
#deck = df['Cabin'].dropna()
df['Embarked'].dropna()
sns.lmplot(x='Age', hue='Sex', y='Survived', data=df)
#sns.lmplot(x='Age', hue='Pclass', y='Survived', data=df)
sns.heatmap(data=df.corr(),annot=True)
```
